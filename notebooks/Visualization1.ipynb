```python
from glob import glob
import os
```

```python
os.chdir("S:\\antpodGIS\\archive\\Excel Sheets")
csvs = glob("APX00000*")
```

```python
import pandas as pd
```

```python
columns = ['meanNDVI', 'meanNDMI']
```

```python
def makeTable(csvs, col):
    os.chdir("S:\\antpodGIS\\kaaviyaTestBench")
    import gisdataframe as gdf
    geo = "FarmKanchi.geojson"
    geodf = gdf.retGeoDF(geo, status=False)
    geodf = geodf.drop('id', axis=1)
    os.chdir("S:\\antpodGIS\\archive\\Excel Sheets")
    for csv in csvs:
        index = pd.read_csv(csv)[col]
        sub = csv.split('-')
        date = [sub[2],sub[3],sub[4]]
        colName = '-'.join(date)
        geodf[colName] = index
    return geodf
```

```python
ndvi = makeTable(csvs, columns[0])
ndvi.head()
```

```python
ndmi = makeTable(csvs, columns[1])
ndmi.head()
```

```python
ndvi.head()
```

```python
def preprocess(geodf):
    if 'id' in geodf.columns:
        geodf = geodf.drop('id', axis=1)
    geodf = geodf.dropna().describe().T
    return geodf[['min', 'max', 'mean']]
```

```python
ndvi = preprocess(ndvi)
ndmi = preprocess(ndmi)
```

```python
ndmi.head()
```

```python
ndvi.head()
```

```python
def forceDate(geodf):
    am = list(geodf.index)
    geodf.reset_index(level=0, inplace=True)
    geodf.rename(columns = {'index':'Date'}, inplace = True) 
    geodf['Date'] = pd.to_datetime(geodf['Date'])
    return geodf
```

```python
ndvi2 = forceDate(ndvi)
ndmi2 = forceDate(ndmi)
```

```python
import matplotlib.pyplot as plt
```

```python
def plotMean(x, y, xlabel, ylabel, title, subplot):
    plt.style.use('seaborn')
    fig = plt.figure(figsize=(10,10))
    ax = plt.axes()
    ax.set_facecolor('#1d262e')
    
    plt.plot(x, y, 'o-', color="#00bb31", linewidth=3, markersize=10)
    
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    
    plt.title(title, fontsize=20)
    plt.legend(['mean'], fancybox=True, shadow=True, loc='best', frameon = True, prop={'size': 12})
    plt.show()
```

```python
index = ['NDVI', 'NDMI']

for i in index:
    plotMean(ndvi2['Date'], ndvi2['mean'], 'Time', i, 'Absolute mean NDVI values from JAN20 to JUL20')
plotMean(ndmi2['Date'], ndmi2['mean'], 'Time', i, 'Absolute mean NDMI values from JAN20 to JUL20')
```

```python
plotMean(ndvi2['Date'], ndvi2['mean'], 'Time', 'NDVI', 'Absolute mean NDVI values from JAN20 to JUL20')
plotMean(ndmi2['Date'], ndmi2['mean'], 'Time', 'NDMI', 'Absolute mean NDMI values from JAN20 to JUL20')
```
