# Raster Analysis

Run this jupyter notebook cell by cell once all the packages listed in setup.txt has been downloaded in the current environment.

```python
import antpodTest as apd
```

#### These are the available products from Sentinel 2 that have been sorted to our need :

1. Sentinel raw bands = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B9', 'B11', 'B12']
2. Sentinel true color with atmospheric correction = [R, G, B]
3. Sentinel VI bands = ['EVI', 'SAVI', 'ARVI']
4. Sentinel VI2 bands = ['NDVI' ,'GNDVI', 'PSRI', 'NDCI']

#### Run configuration parameters to setup loading them:

```python
userID = "APX00000"
mapConfig = "XX"
sensor = ["L2A", "L1C"]
```

A constant configuration parameter - antpodProduct - will mention which of the above 4 products we are trying access.

antpodProduct = one of {  "BANDS", "TC", "VI2", "VI"}

#### We will choose "VI2" to run NDVI analysis.

```python
antpodProduct =  "VXX"
```

Run the below function to get the list of names of all tif product files.

```python
files = apd.rasterCaller(userID, mapConfig, sensor, antpodProduct)
files
```

Run the below function see the metadata of rasters in a dictionary format.

```python
meta = apd.openRaster(files, antpodProduct)
meta
```

```python
numpyListofList = apd.readMultipleDatasetReader(meta)
numpyListofList
```

There are 36 images, hence corresponding to 36 datasets. 

Let's converge our analysis  to one such dataset on a given date.

Extract the list of dates available when the image was taken and choose an arbitrary one:

```python
list_dates = apd.extractDates(meta)
#date = list_dates[32]
#date
```

Extract the dataset that is present in meta on this date:

```python
dataset = meta[date]
dataset
```

Read the tif images inside dataset:

```python
numpyListofBands = apd.readMultipleRaster(dataset)
len(numpyListofBands)
```

It means it has 4 bands in order of VI2 products -  ['NDVI' ,'GNDVI', 'PSRI', 'NDCI']

Extract the NDVI (or change the index to your choice) one to visualize:

```python
ndvi = numpyListofBands[0]
apd.showRaster(ndvi, cmap='RdYlGn')
```

#### Read the shapefile for given farm.

head() returns first 5 rows of a dataframe.

```python
geodf = apd.readShapefile()
geodf
```

```python
geodf = apd.setGeoDFCRS(geodf, epsg=4326)
index = 1
apd.plotShapefileOnRaster(dataset, geodf, index, cmap='RdYlGn')
```

```python
VI2Products =['NDVI' ,'GNDVI', 'PSRI', 'NDCI']
geodf = apd.fillMeanIndexValues(dataset,geodf,VI2Products)
geodf.head()
```

Now let's do the same for prduct VI: 

```python
antpodProduct = "VI"
files2 = apd.rasterCaller(userID, mapConfig, sensor, antpodProduct)
meta2 = apd.openRaster(files2, antpodProduct)
dataset2 = meta2[date]

VIProducts = ['EVI', 'SAVI', 'ARVI'] 
geodf = apd.fillMeanIndexValues(dataset2,geodf,VIProducts)
geodf.head()
```

```python
geodf = apd.insertGeoDFCol(geodf, 'date', date)
geodf.head()
```

```python
products = {"VI2" : ['NDVI' ,'GNDVI', 'PSRI', 'NDCI'], "VI" : ['EVI', 'SAVI', 'ARVI']}

for prod,values in products.items() :
    
    files = apd.rasterCaller(userID, mapConfig, sensor, prod)
    meta = apd.openRaster(files, prod)
    numpyListofList = apd.readMultipleDatasetReader(meta)
    list_dates = apd.extractDates(meta)
    date = list_dates[34]
    #print("Acquistion date: ", date)
    dataset = meta[date]
    geodf = apd.fillMeanIndexValues(dataset,geodf,values)

geodf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

geodf.plot('meanNDVI', figsize = (15,15), cmap = 'RdYlGn',  legend = True)
geodf.boundary.plot(ax=plt.gca(),color='k')
```

```python
products = {'VXX': ['NDVI', 'SAVI', 'AVI', 'EVI', 'ARVI', 'GNDVI', 'NDCI', 'NPCRI', 'PSRI', 'BSI', 'NDMI', 'NDWI', 'SI']}
products
```

```python
for date in list_dates:
    for prod,values in products.items() :
        files = apd.rasterCaller(userID, mapConfig, sensor, prod)
        meta = apd.openRaster(files, prod)
        numpyListofList = apd.readMultipleDatasetReader(meta)
        list_dates = apd.extractDates(meta)
        print("Acquistion date: ", date)
        dataset = meta[date]
        geodf = apd.fillMeanIndexValues(dataset, geodf, values)
        csvFile = "APX-" +"XX1-" +  date + "-MEAN-INDICES.csv"
        geodf.to_csv(csvFile)

geodf
```
