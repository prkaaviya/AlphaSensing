```python
import ee
import sys
sys.path.insert(1, "../apGIS/")
import apgis
from apgis.apgeojson import GeoJSON
from apgis.apfield import Field
import apgis.geespatial as spatial
from apgis.apdate import Date
import apgis.apexception as apexception 
```

```python
localfileName = "f-01"
```

```python
drange = (Date("2020-08-15"), Date("2020-09-30"))
field = Field(localfile=f"{localfileName}.geojson")

S2 = apgis.gee.genCollection(sensor="L2A", daterange=drange, geometry=field.eeROI)\

print(S2.size().getInfo())
```

```python
def algoSCL(image, histlist):
    histlist = ee.List(histlist)
    scl = ee.Image(image.select(["SCL"]).clip(field.eeAOI))
    histo = scl.reduceRegion(ee.Reducer.frequencyHistogram()).get("SCL")
    histlist = histlist.add(histo)
    
    return histlist

x = S2.iterate(algoSCL, ee.List([]))
```

```python
x.getInfo()
```

```python
def colClip(image, imagelist):
    image = image.clip(field.eeAOI)
    imagelist = ee.List(imagelist).add(image)
    return imagelist
```

```python
y = S2.iterate(colClip, ee.List([]))
```

```python
z = ee.ImageCollection.fromImages(y)
```
